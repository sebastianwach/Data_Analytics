```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#1. Normal distribution.

normalDict = {
  "Normal: Mean=0, SD=1": np.random.normal(0, 1, 1000),
    "Normal: Mean=10, SD=1": np.random.normal(10, 1, 1000),
    "Normal: Mean=0, SD=5": np.random.normal(0, 5, 1000)
}

dataframe=pd.DataFrame.from_dict(normalDict)
meltedData=dataframe.melt()


normal=sns.FacetGrid(meltedData,col="variable", sharex=True)
normal.map(plt.hist,"value",cumulative=False, bins=10)

normal=sns.FacetGrid(meltedData,col="variable", sharex=True)
normal.map(plt.hist,"value",cumulative=False, bins=100)

normalCum=sns.FacetGrid(meltedData, col="variable", sharex=True)
normalCum.map(plt.hist,"value",cumulative=True,bins=100)

"""
In normal distribution we can easily see on histogram where is mean value. 
Standard deviation is harder to measure but if this value is higher then histogram is more flat.
If we set too high number of bins then histogram can be much more sharper
Cumulative histogram is much more wider with higher standard deviation.
"""
```

```python
#2. Expit is inverse of logit
from scipy.special import expit
expitDataframe=expit(dataframe)
meltedExpit=expitDataframe.melt()

graphExpit=sns.FacetGrid(meltedExpit, col="variable", sharex=False)
graphExpit.map(plt.hist, "value", cumulative=False, bins=100)
graphExpitCum=sns.FacetGrid(meltedExpit, col="variable", sharex=False)
graphExpitCum.map(plt.hist, "value", cumulative=True, bins=100)

"""
Inverse logit function increase values on edges and normalise these values.
For cumulative histogram: If standard deviation is high and near 0 then histogram is more flat

"""
```

```python
# 3. Poisson distribution.
poissonDict={
    "Poisson: Interval=1":np.random.poisson(1,1000),
    "Poisson: Interval=10":np.random.poisson(10,1000),
    "Poisson: Interval=50":np.random.poisson(50,1000)
}

dataframePoisson=pd.DataFrame.from_dict(poissonDict)
MeltedPoisson=dataframePoisson.melt()

graphPoisson=sns.FacetGrid(MeltedPoisson, col="variable", sharex=True, col_wrap=4)
graphPoisson.map(plt.hist, "value", cumulative=False, bins=100)
graphPoisson_Cum=sns.FacetGrid(MeltedPoisson, col="variable", sharex=True, col_wrap=4)
graphPoisson_Cum.map(plt.hist, "value", cumulative=True, bins=100)

"""
In this case variance and mean is equal to lambda
"""
```

```python
#4. Beta distribution
betaDict={
    "Beta: alpha=1 beta=1":np.random.beta(1,1,1000),
    "Beta: alpha=20 beta=1":np.random.beta(20,1,1000),
    "Beta: alpha=5 beta=5":np.random.beta(5,5,1000)
}
dataframeBeta=pd.DataFrame.from_dict(betaDict)
MeltedBeta=dataframeBeta.melt()

GraphBeta=sns.FacetGrid(MeltedBeta, col="variable", sharex=True, col_wrap=3)
GraphBeta.map(plt.hist, "value", cumulative=False, bins=100)
GraphBetaCum=sns.FacetGrid(MeltedBeta, col="variable", sharex=True, col_wrap=3)
GraphBetaCum.map(plt.hist, "value", cumulative=True, bins=100)

"""
In this case mean is equal to alpha/(alpha+beta)
So if alpha is small histogram goes to the left side and if its vlaue if huge histogram goes to the right side.

"""
```

```python
#5. Log-normal distribution.
lognormalDict={
    "Lognormal: M=0 S=1":np.random.lognormal(0,1,1000),
    "Lognormal: M=5 S=0.5":np.random.lognormal(5,0.5,1000),
    "Lognormal: M=5 S=1":np.random.lognormal(5,1,1000),
    "Lognormal: M=10 S=0.1":np.random.lognormal(10,0.1,1000),
    "Lognormal: M=10 S=0.5":np.random.lognormal(10,0.5,1000),
    "Lognormal: M=10 S=1":np.random.lognormal(10,1,1000)
}

dataframeLognormal=pd.DataFrame.from_dict(lognormalDict)
meltedLognormal=dataframeLognormal.melt()

graphLognormal=sns.FacetGrid(meltedLognormal, col="variable", sharex=False, sharey=False, col_wrap=3)
graphLognormal.map(plt.hist, "value", cumulative=False, bins=100)
graphLognormalCum=sns.FacetGrid(meltedLognormal, col="variable", sharex=False, sharey=False, col_wrap=3)
graphLognormalCum.map(plt.hist, "value", cumulative=True, bins=100)

"""
Mean and variance have exponential values depend from input parameters so values rise really fast 
"""
```

### Lab 3.1
Gamma family distributions. These distributions are defined for positive real numbers. Provide histograms of Gamma and invGamma distributions. See how tail behavior of those distributions behave. Using albebra_solver (see notebook from lecture on poisson distribution modelling) find parameters for Gamma and invGamma for which 98% probability is for values between 5 and 10. Sample for those distributions and compare histograms.

```python
with open('gamma_sampling.stan', 'r') as file:
    print(file.read())
```

```python
gammatuning=pystan.StanModel('gamma_sampling.stan')
```

```python
qua=[10]
data=dict(y_guess=np.array([1]),quantile=np.array(qua))
gammatuned = gammatuning.sampling(data=data,
                          seed=1052020,
                          algorithm="Fixed_param",
                          iter=1, 
                          warmup=0,
                          chains=1)
beta = np.round(gammatuned.extract()['beta'][0])
print(beta)
```

```python
sigma=1
alpha=2
qua=[10]
fig, ax2 = plt.subplots(1, 1,figsize=(7, 4))
x2=np.linspace(0,20,1000)
x4=np.linspace(5,10,1000)

ax2.plot(x2,stats.gamma.pdf(x2,a=alpha,scale=sigma,loc=5),color=dark,linewidth=1)
ax2.fill_between(x4,stats.gamma.pdf(x4,a=alpha,scale=sigma,loc=5),0,color=dark)
ax2.set_yticks([])
ax2.set_xticks([0,5,qua[0]])
plt.show()
```

### Lab 3.2
Negative binomial distribution. Negative binomial distribution can be used in order to relax the constraints of equal mean and variance of Poissson distribution. Using both concentration and dispersion parametrization sample from negative binomial with fixed mean and see how it behaves in comparison to the Poisson with the same mean.


```python
with open('nbinomial.stan', 'r') as file:
    print(file.read())
```

```python
model_prior=pystan.StanModel('globe_prior.stan')
```

```python
R=1000
data=dict(mu=10,phi=10)

sim_uf=model_prior.sampling(data=data,algorithm="Fixed_param", iter=R, warmup=0, chains=1, refresh=R,
                     seed=29042020)
```

```python

mu=10;
data=dict(mu=10,phi=10)
sim_uf=model_prior.sampling(data=data,algorithm="Fixed_param", iter=R, warmup=0, chains=1, refresh=R,
                     seed=29042020)

params=sim_uf.extract()
S_sim=params['S_sim']

fig, axes = plt.subplots(1, 2, figsize=(7, 4))
ax1=axes[0]
ax1.set_yticks([])
ax1.set_title(r'mu=10,phi=10')
ax1.hist(S_sim,color=dark,edgecolor=dark_highlight,density=True,bins=12)
x = np.arange(stats.poisson.ppf(0.0001, mu),stats.poisson.ppf(0.9999, mu))
ax1.plot(x,stats.poisson.pmf(x,mu),color='black',linewidth=1)

#2subplot
data=dict(mu=10,phi=5)

sim_uf=model_prior.sampling(data=data,algorithm="Fixed_param", iter=R, warmup=0, chains=1, refresh=R,
                     seed=29042020)

params=sim_uf.extract()
S_sim=params['S_sim']


ax2=axes[1]
ax2.hist(S_sim,color=dark,edgecolor=dark_highlight,density=True,bins=12)
x = np.arange(stats.poisson.ppf(0.0001, mu),stats.poisson.ppf(0.9999, mu))
ax2.plot(x,stats.poisson.pmf(x,mu),color='black',linewidth=1)
ax2.set_yticks([])
ax2.set_title('mu=10,phi=5')

plt.show()
```

### Summary of lab 3.2:
Yellow histogram is negative bionomial distribution with parameters in title. Black plot is a poisson distribution with mean 10. 
We see that tail on negative binomial distribution is higher than in poisson distribution.


### Lab 3.3
Student-t family distributions. Student-t distributions can be considered generalizations of Gaussian distributions with varying scale. Use both degrees of freedom and dispersion parametrizations to see how changing parameters moves from normal distribution. Take a special case of degrees of freedom equal to 1, as this is Cauchy distribution. See how mean and variance behave while you increase number of samples.


```python
with open('student_t.stan', 'r') as file:
    print(file.read())
```

```python
model_prior=pystan.StanModel('student_t.stan')
```

```python
R=1000
mu=10;
sigma=0.1;
data=dict(nu=1,mu=mu,sigma=sigma)
sim_uf=model_prior.sampling(data=data,algorithm="Fixed_param", iter=R, warmup=0, chains=1, refresh=R,
                     seed=29042020)

params=sim_uf.extract()
S_sim=params['S_sim']

fig, axes = plt.subplots(4, 1, figsize=(5, 13))
ax1=axes[0]
ax1.set_yticks([])
ax1.set_title(r'nu=1,mu=10,sd=0.1')
ax1.hist(S_sim,color=dark,edgecolor=dark_highlight,density=True,bins=100)
x = np.arange(stats.norm.ppf(0.0001,loc=mu,scale=sigma),stats.norm.ppf(0.9999, loc=mu,scale=sigma))
ax1.plot(x,stats.norm.pdf(x,mu,sigma),color='black',linewidth=1)

#2subplot
data=dict(nu=5,mu=10,sigma=2)
sim_uf=model_prior.sampling(data=data,algorithm="Fixed_param", iter=R, warmup=0, chains=1, refresh=R,
                     seed=29042020)
params=sim_uf.extract()
S_sim=params['S_sim']
ax2=axes[1]
ax2.hist(S_sim,color=dark,edgecolor=dark_highlight,density=True,bins=12)
x = np.arange(stats.poisson.ppf(0.0001, mu),stats.poisson.ppf(0.9999, mu))
ax2.plot(x,stats.poisson.pmf(x,mu),color='black',linewidth=1)
ax2.set_yticks([])
ax2.set_title('mu=10,phi=5')

#3subplot
data=dict(nu=11,mu=10,sigma=2)
sim_uf=model_prior.sampling(data=data,algorithm="Fixed_param", iter=R, warmup=0, chains=1, refresh=R,
                     seed=29042020)
params=sim_uf.extract()
S_sim=params['S_sim']
ax3=axes[2]
ax3.hist(S_sim,color=dark,edgecolor=dark_highlight,density=True,bins=12)
x = np.arange(stats.poisson.ppf(0.0001, mu),stats.poisson.ppf(0.9999, mu))
ax3.plot(x,stats.poisson.pmf(x,mu),color='black',linewidth=1)
ax3.set_yticks([])
ax3.set_title('mu=10,phi=5')

#4subplot
data=dict(nu=25,mu=10,sigma=2)
sim_uf=model_prior.sampling(data=data,algorithm="Fixed_param", iter=R, warmup=0, chains=1, refresh=R,
                     seed=29042020)
params=sim_uf.extract()
S_sim=params['S_sim']
ax4=axes[3]
ax4.hist(S_sim,color=dark,edgecolor=dark_highlight,density=True,bins=12)
x = np.arange(stats.poisson.ppf(0.0001, mu),stats.poisson.ppf(0.9999, mu))
ax4.plot(x,stats.poisson.pmf(x,mu),color='black',linewidth=1)
ax4.set_yticks([])
ax4.set_title('mu=10,phi=5')


plt.show()
```

### Summary to lab 3.3:

For df=1 it is a cauchy distribution. Where most of prior is concentrate in one value.
Increasing degrees freedom  makes pdf more flat and wider.
